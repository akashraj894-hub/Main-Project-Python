# Dataset Information

Source: https://www.kaggle.com/datasets/txtpwsss/dirty-financial-risk-dataset

Year / Timeline: Data collected during 2023

Dataset Description:

The dataset contains financial transaction records of bank customers. It includes details such as customer information, transaction data, credit details, and market indicators used for financial analysis and risk assessment.

Name : It represents the name of the customer.

Age : It indicates the age of the customer.

Gender : It specifies the gender of the customer.

Account Balance : It shows the available balance in the customer’s account.

Transaction Amount : It represents the amount involved in each transaction.

Transaction Type : It indicates whether the transaction is a credit or debit.

Credit Score : It represents the creditworthiness of the customer.

Risk Level : It classifies customers into low, medium, or high risk.

Loan Amount : It indicates the loan amount taken by the customer.

Interest Rate : It shows the interest rate applied to the loan or financial product.

Inflation Rate : It represents the economic inflation rate at the time of transaction.

Market Volatility Index : It indicates the level of market fluctuations.

Timestamp : It shows the date and time of the transaction.

Device Type : It represents the device used for the transaction (mobile/desktop).

Location : It indicates the customer’s geographical location.

# Type of Analysis

**Descriptive Analysis (summarizing the dataset)**

In this project, descriptive analysis was performed to summarize the financial dataset. We analyzed key variables such as transaction amount, account balance, credit score, and risk level. Data was grouped based on transaction type and risk level to understand distribution patterns. We also examined average values of market indicators like interest rate, inflation rate, and market volatility index. This helped identify common transaction behaviors, customer risk segments, and overall financial trends.

**Diagnostic Analysis (finding reasons behind patterns)**

Diagnostic analysis was conducted to identify factors influencing financial risk and transaction behavior. We analyzed the relationship between credit score and risk level, transaction amount and account balance, and market volatility with investment duration. Additionally, we examined how device usage metrics (CPU and memory usage) relate to user activity patterns. This helped identify key drivers affecting risk classification and customer financial behavior.

**Predictive Analysis (optional if forecasting required using basic stats)**

Predictive analysis was performed by analyzing trends over time using the timestamp column. We studied patterns in transaction amounts and risk levels across different time periods. Based on historical data, we identified periods with higher financial activity and potential risk fluctuations. This helps in forecasting future transaction trends and anticipating high-risk scenarios

**Prescriptive Analysis (recommendations for business decisions)**

Based on the analysis, several recommendations were provided. It is suggested to closely monitor users with low credit scores and high transaction amounts, as they may indicate higher risk. Financial institutions can optimize decision-making by considering market indicators like inflation and interest rates. Additionally, improving monitoring systems using device usage patterns can help detect unusual activities. Businesses can also implement targeted strategies for different risk-level customers to improve financial stability and reduce potential losses.

## Stage 2 – Data Cleaning and Pre-processing

Handle missing values (impute or drop)

Handle duplicates

Treat outliers if required

```python
df
```

To remove duplicates

```python
df = df.drop_duplicates(subset='user_id', keep='first')
df
```

To remove USD in transaction_amount

```python
df['transaction_amount'] = df['transaction_amount'].replace('[USD]', '', regex=True)
df
```

To change the Amount coloumn in numeric

```python
numeric_cols = ['transaction_amount','account_balance','market_volatility_index',
                'interest_rate','inflation_rate','stock_index_change',
                'investment_duration_days','credit_score',
                'device_cpu_usage','device_memory_usage']
df
```

To update the missing values in transaction_amount

To check the null

```python
df['transaction_amount'].isnull().sum()
```

```python
df['account_balance'] = df['account_balance'].replace('[$]', '', regex=True)
df
```

To change the decimals

```python
cols = ['transaction_amount','account_balance','market_volatility_index',
        'interest_rate','inflation_rate','stock_index_change',
        'investment_duration_days','credit_score',
        'device_cpu_usage','device_memory_usage']

df[cols] = df[cols].apply(lambda x: x.round(2))

df
```

To change the incosistance in transaction_type

```python
wrong_values = [
    'wdrialwhat','wlaridhtwa','lhdwraatwi','wldaithawr',
    'lhirtdwaaw','awwdilhart','widatrlhwa','aarwdtliwh','hltwaiawrd'
]

df.loc[df['transaction_type'].isin(wrong_values), 'transaction_type'] = 'Withdrawal'

df
```

```python
wrong_values = [
    'dopeits','topieds','poeitsd','eidstop', 'pditseo'

]

df.loc[df['transaction_type'].isin(wrong_values), 'transaction_type'] = 'Deposit'

df
```

```python
wrong_values = [
    'rsfenart','rsnaertf','esrtfarn','srfeatnr', 'eafnrstr','etfasnrr','tsrfenra'

]

df.loc[df['transaction_type'].isin(wrong_values), 'transaction_type'] = 'Transfer'

df
```

```python
wrong_values = [
    'meievtnsnt','vemtneistn','tneemsvnit','vsmnitetne', 'evinemntst','neminesvtt','eitvetnsnm','INVESTMENT'

]

df.loc[df['transaction_type'].isin(wrong_values), 'transaction_type'] = 'Investment'

df
```

To replace the with the help of mode in transaction type

```python
mode_val = df['transaction_type'].mode()[0]
df['transaction_type'] = df['transaction_type'].replace('INVALID_CATEGORY_DEFAULT', mode_val)

df
```

To Change the values as upper

```python
df['transaction_type'] = df['transaction_type'].str.upper()
df
```

To change the u as U and remove the - in user_id

```python
df['user_id'] = df['user_id'].str.upper()
df['user_id'] = df['user_id'].str.replace('-', '', regex=False)
df
```

```python
df

```

Corrected the incositance in timestamp column and seprated the date and time

```python
df['timestamp'] = pd.to_datetime(df['timestamp'], errors='coerce')
mode_time = df['timestamp'].mode()[0]
df['timestamp'] = df['timestamp'].fillna(mode_time)

df
```

```python
df['Date'] = df['timestamp'].dt.date
df['Time'] = df['timestamp'].dt.time
df
```

```python
cols = ['Date','Time'] + [col for col in df.columns if col not in ['Date','Time']]
df = df[cols]
```

```python
df
```

Corrected the proper format in heading

```python
df.columns = df.columns.str.title()
df.head()
```

```python
df.to_csv('cleaned_datafinalset.csv', index=False)
```
